```python
import yfinance as yf
```

```python
spy = yf.Ticker("SPY")
option_expiries = spy.options
```

```python
print(option_expiries)
```

```python
from google.colab import data_table
data_table.enable_dataframe_formatter()
```

```python
call_options = spy.option_chain(option_expiries[1])
data_table.DataTable(call_options.calls, include_index=False, num_rows_per_page=30)
```

```python
put_options = spy.option_chain(option_expiries[1])
data_table.DataTable(put_options.puts, include_index=False, num_rows_per_page=30)
```
